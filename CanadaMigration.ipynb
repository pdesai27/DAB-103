## Analysis of country, industry and skill migration for Canada

### Country

```python
# Canadian Migration (in & out)
df_canada = df[df.base_country_name == 'Canada']
df_canada = df_canada.reset_index(drop=True)
```

```python
# Use head to display first five rows
df_canada.head(5)
```

```python
# Drop unwanted columns
df_canada = df_canada.drop(columns=['target_lat','target_long', 'base_lat', 'base_long'])
```

```python
# Use head to display first five rows
df_canada.head(5)
```

```python
# Calculate average migration value across the row
df_canada['avg_country_migration'] = df_canada.mean(axis=1)

# Sort by average migration value in descending order
df_canada.sort_values(by='avg_country_migration',ascending=False)
```

```python
# Create csv file for Canadian migration and use it to create visualization in Tableau
df_canada.to_csv('df_canada.csv', header=True, index=False)
```

### Industry

```python
#Canadian Industry dataframe (in & out)
df2_canada = df2[df2.country_name == 'Canada']
df2_canada = df2_canada.reset_index(drop=True)
```

```python
# Use head to display first five rows
df2_canada.head(5)
```

```python
# Drop unwanted columns
df2_canada = df2_canada.drop(columns = ['isic_section_index','industry_id'])
```

```python
# Use head to display first five rows
df2_canada.head(5)
```

```python
# Calculate average migration value across the row
df2_canada['avg_industry_migration'] = df2_canada.mean(axis=1)

# Sort by average migration value in descending order
df2_canada.sort_values(by='avg_industry_migration',ascending=False)
```

```python
# Create csv file for Canadian industries and use it to create visualization in Tableau
df2_canada.to_csv('df2_canada.csv', header=True, index=False)
```

### Skill
